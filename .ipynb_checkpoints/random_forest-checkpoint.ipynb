## Random Forest
In this file, Random Forest model is fit by tuning different hyperparameters based on F_{1.5} score: Criterion for splitting, minimum number of samples to split, and maximum number of features for choosing to split.<br>
We will look at the *slopes* of precision and recall across different hyperparameters to see its influence rather than their face-values since the cut-off probability is set to 0.5 for the tuning process before deciding the threshold probability.<br>
Also, by comparing the fit to the training set vs. validation set, it is easy to see how trees tend to overfit. Lastly, the cut-off probability is decided by the F score.

Import the modules

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from ipynb.fs.full.helper_functions import * # Custom function to create dummy variables
from sklearn.metrics import precision_recall_fscore_support, f1_score, accuracy_score, roc_auc_score

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

# Make better use of Jupyter Notebook cell width
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
```

Import the imputed training set and validation set

```python
X_train, X_val, y_train, y_val = pd.read_pickle('data/imp_train.pkl'), pd.read_pickle('data/imp_val.pkl'), pd.read_pickle('data/y_train.pkl'), pd.read_pickle('data/y_val.pkl')
```

### Hyperparameter Tuning :<br>
1. criterion = {“gini”, “entropy”} <br>
2. min_samples_split from 5 to 40(the minimum number of samples required to split an internal node)<br>
3. max_features from 8 to 30(the maximum number of features to choose from when splitting)<br>
4. n_estimators from 20 to 30(the number of trees)

- Remark: Only min_samples_split has a heavy effect on precision and little effect on recall. The other hyperparameters have little effect on both precision and recall.

#### 1. Criterion

Criterion = Gini<br>
F score is almost 1.0 in the training fit, but 0.5 in the validation fit. This is overfit.

```python
print(f'Class weight for positive label: {sum(y_train == 0) // sum(y_train == 1)}')
randomforest = RandomForestClassifier(n_estimators=100, class_weight = {0: 1, 1: 27}, oob_score = True, random_state= 10,
                                      criterion = "gini")
_, scores_train, scores_val = randomforest_result(randomforest, X_train, y_train, X_val, y_val);
```

Entropy has slightly better F score(0.51) than Gini

```python
randomforest_ent = RandomForestClassifier(n_estimators=100, class_weight = {0: 1, 1: 27}, oob_score = True, random_state= 10,
                                          criterion = "entropy")
_, scores_train, scores_val = randomforest_result(randomforest_ent, X_train, y_train, X_val, y_val);
```

#### 2. min_samples_split from 5 to 40

```python
# Read the saved scores lists
scores_train_min = pd.read_pickle('random_forest/scores_train_min_samples_split.pkl')
scores_val_min = pd.read_pickle('random_forest/scores_val_min_samples_split.pkl')
```

```python
scores_train_min, scores_val_min = [], [] 
min_samples_split_list = np.arange(5, 41)

for min_samples_split in min_samples_split_list:
    print(f"min_samples_split = {min_samples_split}")
    randomforest_ent = RandomForestClassifier(n_estimators=100, class_weight = {0: 1, 1: 27}, oob_score = True, random_state= 10,
                                              criterion = "entropy",
                                              min_samples_split = min_samples_split)
    _, scores_train_min, scores_val_min = randomforest_result(randomforest_ent, X_train, y_train, X_val, y_val, scores_train_min, scores_val_min)
```

Fit to the training set
- As the minium sample size increases(trees getting shorter), precision drops a bit faster than recall. So, it would mean that there are more false positives as the tree gets shorter. 

```python
# Fit to the training set
plot_metric_min_samples_split(min_samples_split_list, scores_train_min)
```

Fit to the validation set
- But when it comes to generalization, the number of true positives slowly increases(recall increasing) as the trees get shorter.
- 38 minimum sample size has the best F score.

```python
# Fit to the validation set
plot_metric_min_samples_split(min_samples_split_list, scores_val_min)
```

#### 3. max_features from 8 to 30

```python
# Read the saved scores lists
scores_train_max = pd.read_pickle('random_forest/scores_train_max_features.pkl')
scores_val_max = pd.read_pickle('random_forest/scores_val_max_features.pkl')
```

```python
scores_train_max, scores_val_max = [], []
max_features_list = range(8, 31)
for max_features in max_features_list:
    print(f"max_features = {max_features}")
    randomforest_ent = RandomForestClassifier(n_estimators=100, class_weight = {0: 1, 1: 27}, oob_score = True, random_state= 10,
                                              criterion = "entropy",
                                              min_samples_split = 38,
                                              max_features = max_features)
    _, scores_train_max, scores_val_max = randomforest_result(randomforest_ent, X_train, y_train, X_val, y_val, scores_train_max, scores_val_max)
```

Fit to the training set
- Maximum number of features doesn't have a lot of influence on recall ratio which is probably because the probabilities are over estimated due to class weight hyperparameter. 
- As more features become available, the trees would slightly better classify the examples(precision increasing slightly).

```python
# Fit to the training set
plot_metric_max_features(max_features_list, scores_train_max)
```

Fit to the validation set
- For unseen dataset as well, precision and recall are hardly influenced by increasing the number of features from 8 to 30

```python
# Fit to the validation set
plot_metric_max_features(max_features_list, scores_val_max)
```

#### 4. n_estimators from 20 to 30

```python
# Read the saved scores lists
scores_train_n_estimators = pd.read_pickle('random_forest/scores_train_n_estimators.pkl')
scores_val_n_estimators = pd.read_pickle('random_forest/scores_val_n_estimators.pkl')
```

```python
scores_train_n_estimators, scores_val_n_estimators = [], []
for n_estimators in range(20, 31):
    print(f"n_estimators = {n_estimators}")
    randomforest_ent = RandomForestClassifier(n_estimators=n_estimators, class_weight = {0: 1, 1: 27}, oob_score = True, random_state= 10,
                                              criterion = "entropy",
                                              min_samples_split = 38,
                                              max_features = 'sqrt')
    _, scores_train_n_estimators, scores_val_n_estimators = randomforest_result(randomforest_ent, X_train, y_train, X_val, y_val, scores_train_n_estimators, scores_val_n_estimators)
```

Fit to the training set
- Increasing the number of trees from 20 to 30 does not seem to influence precision nor recall. This would probably mean that many trees are similar to each other as if they are duplicated so the decision doesn't change.

```python
# Fit to the training set
plot_metric_n_estimators(range(20, 31), scores_train_n_estimators)
```

Fit to the validation set
- For the unseen data as well, the number of trees doesn't have add a lot of meaning.

```python
# Fit to the validation set
plot_metric_n_estimators(range(20, 31), scores_val_n_estimators)
```

### Final Random Forest Model is...

Final results: F_{1.5} = **0.61** with precision = 0.56 and recall = 0.64 on the validation set.

```python
randomforest_final = RandomForestClassifier(class_weight = {0: 1, 1: 27}, 
                                            random_state= 10,
                                            criterion = "entropy",
                                            n_estimators=30,
                                            min_samples_split = 38,
                                            max_features = 'sqrt')

randomforest_final.fit(X_train, y_train);
y_probs = randomforest_final.predict_proba(X_val)[:, 1]
precs, recs, fbetas, acc_scores = plot_fbeta_recall_precision_acc(y_probs, y_val)
```

Fit to the training set to check overfitting

```python
# Fit to the training set to check overfitting
y_probs_train = randomforest_final.predict_proba(X_train)[:, 1]
precs, recs, fbetas, acc_scores = plot_fbeta_recall_precision_acc(y_probs_train, y_train)
```

Cut-off probability to determine "Fraud"

```python
# Cut-off probability to determine "Fraud"
threshold = 0.48459
```

```python
pred_train = (randomforest_final.predict_proba(X_train)[:, 1] >= threshold).astype(int)
prec, rec, fbeta, _ = precision_recall_fscore_support(y_train, pred_train , beta = 1.5, average = 'binary')
print(f"Train set: Precision = {round(prec, 5)}, Recall = {round(rec, 5)}, F_1.5 = {round(fbeta, 5)}")

pred_val = (randomforest_final.predict_proba(X_val)[:, 1] >= threshold).astype(int)
prec, rec, fbeta, _ = precision_recall_fscore_support(y_val, pred_val, beta = 1.5, average = 'binary')
print(f"Val set: Precision = {round(prec, 5)}, Recall = {round(rec, 5)}, F_1.5 = {round(fbeta, 5)}\n")
```
