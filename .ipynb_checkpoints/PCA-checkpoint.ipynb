Seems that 1 PC for 'C' variable, 1 PC for 'D' variable, 2 PC's for 'V' variable will do.

```python
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
ax[0].bar(range(1,len(pca_C.components_)+1), pca_C.explained_variance_ratio_)
ax[0].set_xticks(range(1, len(pca_C.components_)+1))
ax[0].set_xlabel('PC')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].bar(range(1,len(pca_D.components_)+1), pca_D.explained_variance_ratio_)
ax[1].set_xticks(range(1, len(pca_D.components_)+1))
ax[1].set_xlabel('PC')
ax[1].set_ylabel('Explained Variance Ratio')

ax[2].bar(range(1,11), pca_V.explained_variance_ratio_[:10])
ax[2].set_xticks(range(1, 11))
ax[2].set_xlabel('PC')
ax[2].set_ylabel('Explained Variance Ratio');
```

#### Using the same pca, perform PCA on imputed numeric validation set

```python
imp_val = pd.read_pickle('data/imp_val.pkl')
imp_num_val = imp_train.loc[:, :'id_11']
imp_cat_val = imp_train.loc[:, 'C':]

std_C_val = scaler_C.transform(imp_num_val.loc[:, imp_num_val.columns.map(lambda x: x[0]) == 'C'])
std_D_val = scaler_D.transform(imp_num_val.loc[:, imp_num_val.columns.map(lambda x: x[0]) == 'D'])
std_V_val = scaler_V.transform(imp_num_val.loc[:, imp_num_val.columns.map(lambda x: x[0]) == 'V'])

principalComponents_val = pd.DataFrame(np.concatenate([pca_C.transform(std_C_val)[:,0].reshape(-1,1),
                                                       pca_D.transform(std_D_val)[:,0].reshape(-1,1), 
                                                       pca_V.transform(std_V_val)[:,:2]], axis =1), columns = ['C_PC1', 'D_PC1', 'V_PC1', 'V_PC2'])
other_cols_val = imp_num_val.loc[:, imp_num_val.columns.map(lambda x: (x[0] != 'C') & (x[0] != 'D') & (x[0] != 'V'))]
PC_num_val = pd.concat([principalComponents_val, other_cols_val], axis = 1)
PC_num_val.head()
```

Merge the PC_num_val and imputed categorical validation data

```python
X_val_pca = pd.concat([PC_num_val, imp_cat_val], axis = 1)
```
